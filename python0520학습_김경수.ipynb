```python
def iris_stat(species, part):
    temp = iris[iris['species'] == species] # indexing
    iris_max = temp[part].max()
    iris_min = temp[part].min()
    iris_mean = temp[part].mean()
    iris_median = temp[part].median()
    
    return {'max':iris_max, 'min':iris_min, 'mean':iris_mean, 'median':iris_median}    # max: 딕셔너리화로 묶어준다.
```

```python
sns.histplot(data=seller, x='대분야', color='darkorange', alpha=.5)
#sns.histplot(data=seller, x='sepal_length', color='lightgreen', alpha=.5)
plt.xlabel('대분야')
plt.ylabel('책 수량(권)')
plt.title('분야별 책 보유 현황')
```

```python
# 갯수체크한후 많은지 적은지 나눠줌
df = seller.groupby('대분야').count()
df['재고'] = df['월'] > df['월'].mean()
df['재고'] = df['재고'].apply(lambda x: '많음' if x else '적음')
#df.head()
```

```python
#df
#df['월'] > df['월'].mean()
#df['재고'].apply(lambda x: '많음' if x else '적음')
df['재고']
```

```python
fig = px.bar(data_frame=df, x=df.index, y='월',
            labels={'대분야','분야','월:책 수량 (권)'},
            title='분야별 책 보유 현황',
            color='재고')
fig.show()
```

```python
# sort_values = sorting하고 시각화 value하는 방법
df.sort_values('재고', inplace=True, ascending=False) # 수량권수가 많은게 오른쪽으로 보여지게 할려면 ascending=False

#적은거 부터 많게 정렬
df.sort_values('월', inplace=True) # 반대로 많은거 -> 적은거 까지 ascending=False 추가
```

```python
#seaborn 방법
plt.subplots(figsize=(30,6))
sns.barplot(data=df, x=df.index, y='월', hue='재고')
plt.xlabel('분야')
plt.ylabel('책 수량 (권)')
plt.title('분야별 책 보유 현황')
plt.xticks(rotation=90)
plt.show()
# 책재고 많음 적음이 분류가안됨 sort value 사용해서 제거할수 있음.
```
