```python
for c in temp.columns[:-1]:
    mean = temp[c].mean()
    print(f'iris {c} 의 평균값 : {mean}')
    
    max_value = temp[c].max()
    print(f'iris {c} 의 최댓값 : {max_value}')
    
    min_value = temp[c].min()
    print(f'iris {c} 의 평균값 : {min_value}')
    
    median = temp[c].median()
    print(f'iris {c} 의 평균값 : {median}')
```

```python
for s in iris['species'].unique():
    temp = iris[iris['species'] == s]
    
    for c in temp.columns[:-1]:
        mean = temp[c].mean()
        print(f'iris {s} {c} 의 평균값 : {mean}')

        max_value = temp[c].max()
        print(f'iris {s} {c} 의 최댓값 : {max_value}')

        min_value = temp[c].min()
        print(f'iris {s} {c} 의 평균값 : {min_value}')

        median = temp[c].median()
        print(f'iris {s} {c} 의 평균값 : {median}')
```

```python
def iris_function(species, part, stepinfo):
    temp = iris[iris['species'] == species]
    temp2 = temp[part]
    
    if stepinfo == 'mean':
        return temp2.mean()
    elif stepinfo == 'median':
        return temp2.median()
    elif stepinfo == 'max':
        return temp2.max()
    elif stepinfo == 'mix':
        return temp2.mix()
```
