```python
import cv2
import matplotlib.pyplot as plt
import urllib
import numpy as np
```

```python
# byte는 2진수라서 array를 numpy로 만들수 있다.
req = urllib.request.urlopen('https://search.pstatic.net/sunny/?src=https%3A%2F%2Fimage.utoimage.com%2Fpreview%2Fcp868526%2F2016%2F11%2F201611023141_800.jpg&type=a340')
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
img = cv2.imdecode(arr, -1)
```

```python
plt.imshow(img)
```

```python
from bs4 import BeautifulSoup
import requests

# headers를 딕셔너리로 정의
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)' 
           'AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0' 
           'Safari/537.36'} #
url = f'https://www.melon.com/chart/index.htm'
content = requests.get(url, headers = headers).content
```

```python
# BeautifulSoup을 사용하여 HTML 파싱
soup = BeautifulSoup(content, 'html.parser')
```
