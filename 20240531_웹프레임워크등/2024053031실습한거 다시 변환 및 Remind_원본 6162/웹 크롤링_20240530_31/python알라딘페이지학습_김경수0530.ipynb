```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#page = 1
result = pd.DataFrame()    # result 데이터프레임으로 만든다

for page in range(1,21):  # #@! 다른방법 : while True: # 몇쪽까지 있을지 모를때 while로
    
    # url에서 page = 1 -> page=[page] 페이지가 나온 부분을 포맷팅 합니다
    url = f'https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page=[page]&cnt=1000&SortOrder=1'

    content = requests.get(url=url).content
    
    soup = BeautifulSoup(content, 'html.parser')

      # 특정 텍스트가 포함된 경우 루프 종료
#     if '요청하신 작업을 수행하지 못했습니다.' in soup.text:
#         break

    title = [s.text for s in soup.select('a.bo3 b')]                                         # 제목
    discount_price = [s.text for s in soup.select('div.ss_book_list ul li span.ss_p2')]      # 할인가 부분
    price = []                 # 판매가
    author = []                # 저자 
    publisher = []             # 출판사
    date = []                  # 출간일
    star = []                  # 별점

    for s in soup.select('div.ss_book_list ul li'): # [s.text for s in soup.select('div.ss_book_list ul li span.ss_p2')]정가 부분은   위에 할인가랑 다르다 클래스 부분이 없다.
        if '원 →' in str(s):
            price.append(s.text.split(' → ')[0])      #  → 기준으로 

    for s in soup.select('div.ss_book_list ul li'):
        if '년' in s.text and '월' in s.text and '|' in s.text:               # 위 에서 and '|' in s.text 플러스 in // #@! s.text 빼도 같은지
            author.append(s.text.split(' | ')[0])              # for i in 으로 되고 [0]으로해도 된다
            publisher.append(s.text.split(' | ')[1])
            date.append(s.text.split(' | ')[2])

    for s in soup.select('div.ss_book_list'):    # div 태그 안에 ss_book_list 클래스
        if '|' in str(s):
            s2 = s.select('span.star_score')   # 빈리스트 인덱싱이 안되서 텍스트로 못씀.

            if s2 != []:
                star.append(s2[0].text)
            else:
                star.append(float('nan'))
                
    urls = [s['href'] for s in soup.select('a.bo3')]   # URL
    
    df_temp = pd.DataFrame([title, price, discount_price, author, publisher, date, star, urls]).T # .T 행과 열을 바꾼다
    result = pd.concat([result, df_temp], axis=0)           

result.columns = ['제목','정가','판매가','저자','출판사','출간일','별점','URL']
```
