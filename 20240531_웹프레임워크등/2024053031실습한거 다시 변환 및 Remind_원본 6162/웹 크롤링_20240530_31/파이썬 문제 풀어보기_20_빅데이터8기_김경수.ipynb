문제 2: a = 55.994이고, b = -75일 때, a를 일의 자릿수까지 반올림한 숫자와, b의 절댓값 중 큰 것을 구하라

```python
# 문제 2 답
import math

a = 55.994
b = -75

rounded_a = round(a)

abs_b = abs(b)   # 절댓값

# 둘 중 큰 값
result = max(rounded_a, abs_b)

print(result)
```

```python
# 문제 8 답
tt_sum = 0

# 범위 내에서 홀수인 x의 합 계산
for x in range(101, 806):
    if x % 2 != 0:
        tt_sum += x

print("100 < x < 806 일때 x의 홀수의 합:", tt_sum)
```
