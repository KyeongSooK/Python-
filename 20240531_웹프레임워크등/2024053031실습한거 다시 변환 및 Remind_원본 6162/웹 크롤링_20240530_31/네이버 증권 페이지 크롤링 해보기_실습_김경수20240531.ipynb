```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

# 사용자로부터 sosok 값을 입력받음
sosok = input('Tap에 원하는 숫자를 치세요(1.코스피 2.코스닥 3.코스피(닥))')

# URL 설정
url = f'https://finance.naver.com/sise/sise_quant.naver?sosok={int(sosok) - 1}'

# 웹 페이지 내용 가져오기
content = requests.get(url).content

# BeautifulSoup을 사용하여 HTML 파싱
soup = BeautifulSoup(content, 'html.parser')

# HTML 테이블을 판다스 데이터프레임으로 변환
table = pd.read_html(str(soup.select('table')[1]))[0]

# 테이블을 정리
if 'N' in table.columns:                        # table DataFrame의 열에 N이 있을 때.
    table.set_index('N', inplace=True)          # N column [0] 인덱스 삭제
    table.dropna(how='all', inplace=True)       # table의 VALUES가 NaN 값인 부분만 모두 삭제

# sosok 값에 따라 테이블을 선택하거나 병합된 테이블을 보여줌
if sosok == '1':
    print("코스피 테이블:")
    print(table)
elif sosok == '2':
    print("코스닥 테이블:")
    print(table)
else:
    print("종합 테이블:")
    # 다른 sosok 값에 대해서 병합된 테이블을 보여줌 (예: 코스피와 코스닥 같이 나옴 병합)
    url2 = f'https://finance.naver.com/sise/sise_quant.naver?sosok=1'
    content2 = requests.get(url2).content
    soup2 = BeautifulSoup(content2, 'html.parser')
    table2 = pd.read_html(str(soup2.select('table')[1]))[0]

    # 테이블을 정리
    if 'N' in table2.columns:
        table2.set_index('N', inplace=True)
        table2.dropna(how='all', inplace=True)

    # 두 테이블 병합
    merged_table = pd.concat([table, table2], ignore_index=True)
    print(merged_table)
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

# 함수 정의: 웹 페이지에서 테이블을 가져오는 함수
def get_table(sosok):
    url = f'https://finance.naver.com/sise/sise_quant.naver?sosok={sosok}'
    content = requests.get(url).content
    soup = BeautifulSoup(content, 'html.parser')
    table = pd.read_html(str(soup.select('table')[1]))[0]
    if 'N' in table.columns:
        table.set_index('N', inplace=True)
    table.dropna(how='all', inplace=True)
    return table

# 사용자로부터 sosok 값을 입력받음
while True:
    sosok = input('Tap에 원하는 숫자를 치세요(1.코스피 2.코스닥 3.exit): ')
    if sosok not in ['1', '2', '3']:
        print("잘못된 입력입니다. 다시 입력하세요.")
        continue
    sosok = int(sosok) - 1
    break

# sosok 값에 따라 적절한 테이블을 가져옴
if sosok == 0:
    table_kospi = get_table(0)
    print("코스피 테이블:")
    display(table_kospi)  # Jupyter Notebook 환경에서는 display 함수 사용
elif sosok == 1:
    table_kosdaq = get_table(1)
    print("코스닥 테이블:")
    display(table_kosdaq)  # Jupyter Notebook 환경에서는 display 함수 사용
else:
    table_kospi = get_table(0)
    table_kosdaq = get_table(1)
    merged_table = pd.concat([table_kospi, table_kosdaq], ignore_index=True)
    print("코스피와 코스닥 병합된 테이블:")
    display(merged_table)  # Jupyter Notebook 환경에서는 display 함수 사용

# 아래 코드는 Jupyter Notebook 환경이 아닌 경우에 DataFrame을 표 형식으로 출력
try:
    from IPython.display import display  # Jupyter Notebook 환경에서 사용
except ImportError:
    def display(df):
        with pd.option_context('display.max_rows', None, 'display.max_columns', None):
            print(df)

# 표 형식으로 출력
if sosok == 0:
    display(table_kospi)
elif sosok == 1:
    display(table_kosdaq)
else:
    display(merged_table)
```
