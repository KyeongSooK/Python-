1. sklearn.Linear_model.SGDRegressor(loss=“squared_loss”, ...)를 사용해서, 5개 특성 중
4개를 인풋으로 하고, 1개 특성을 타겟으로 선택해 연속형 데이터를 예측하는 SGD 회귀분석
모델을 훈련시키고 시각화해 보세요. loss를 제외한 모든 패러미터는 SGDClassifier와 동일합
니다.

```python
import pandas as pd

fish = pd.read_csv('https://bit.ly/fish_csv_data')
fish.head()
```

```python
fish_input = fish[['Weight', 'Length', 'Height', 'Diagonal']].to_numpy()
fish_target = fish['Width'].to_numpy()
```

```python
# 트레인데이터셋 평가데이터셋 나눠줍니다
from sklearn.model_selection import train_test_split
# 기준은 트레인데이터가 되어야한다.
train_input, test_input, train_target, test_target = train_test_split(
    fish_input, fish_target, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
# 싸이키런에 sgd 인포트 해준다
from sklearn.linear_model import SGDRegressor
```

```python
sc = SGDRegressor(loss='squared_error', max_iter=10, random_state=42)
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

```python
import numpy as np

sc = SGDRegressor(loss='squared_error', random_state=42)

train_score = []
test_score = []

classes = np.unique(train_target)
```

```python
for _ in range(0, 300):
    sc.partial_fit(train_scaled, train_target)#, classes=classes)

    train_score.append(sc.score(train_scaled, train_target))
    test_score.append(sc.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt

plt.plot(train_score)
plt.plot(test_score)
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

3. pickle 패키지를 사용하면, 모델링한 가중치를 저장하고 불러와 사용할 수 있습니다.
여러분의 데이터를 5분할(또는 그 이상)하여 txt 또는 csv로 저장하고, partial_fit을 사용
하여 점진학습을 실습해 봅시다. 사용법은 다음과 같습니다.

```python
# 점진적으로 쓸수없으니 해당 모델 가중치를 저장하는 기능
# with쪽 fc = f로 저장 // sc = model 
# 앞에를 모델 sc를 끌어와도 된다!?!@!
```

```python
import numpy as np
from sklearn.linear_model import SGDRegressor
import pickle
```

```python
# 모델 지정
sc = SGDRegressor(max_iter=1000)
sc.fit(train_scaled, train_target)

# 모델 가중치 저장
with open('model_heartetc.pkl', 'wb') as fc:
    pickle.dump(sc, fc)

# model_heartetc.pkl 파일안에는 fit, score, coef_, intercept_ 등의 해당 모델의 결과값이 txt파일로 적혀있다.
```

```python
# 2-1. 2-2 하기 전에 데이터 만들기 ( numpy패키지의 to_numpy 로 된 데이터를 csv로 저장 )

# NumPy 배열 생성

# NumPy 배열을 DataFrame으로 변환
#df = pd.DataFrame(heart_input, columns=['age','trtbps','chol','thalachh'])
#df1 = pd.DataFrame(heart_target, columns=['output'])

df = heart[['age','trtbps','chol','thalachh']]
df1 = heart[['output']]

# DataFrame을 CSV 파일로 저장
df.to_csv('heart_split_input.csv', index=False)
df1.to_csv('heart_split_target.csv', index=False)

print("CSV 파일이 'heart_split_input.csv , heart_split_target.csv' 두개의 이름으로 저장되었습니다.")
```

```python
# 2-2. 새로운 데이터셋으로 점진학습
# 새로운 데이터셋 로드
input_partial = np.array(pd.read_csv('heart_split_input.csv')) # heart[['age','trtbps','chol','thalachh']]
target_partial = np.array(pd.read_csv('heart_split_target.csv')).reshape([-1,1]) # heart[['output']]

#모델 로드
with open('model_heartetc.pkl', 'rb') as fc:
    sc = pickle.load(fc)

#점진학습
sc.partial_fit(input_partial, target_partial)

#업데이트된 모델 가중치 저장
with open('model_heartetc.pkl', 'wb') as fc:
    pickle.dump(sc, fc)
```

```python
import pickle

# 모델 로드
with open('model_heartetc.pkl', 'rb') as fc:
    model = pickle.load(fc)

# 모델의 타입 출력
print(f"모델 타입: {type(model)}")
```

```python

# 모델의 하이퍼파라미터 확인 (예: scikit-learn 모델의 경우)
if hasattr(model, 'get_params'):
    print("모델 하이퍼파라미터:")
    for param, value in model.get_params().items():
        print(f"  {param}: {value}")
```

```python

# 모델의 학습된 파라미터 확인 (예: scikit-learn 모델의 경우)
if hasattr(model, 'coef_'):
    print(f"모델의 가중치 (coef_): {model.coef_}")
```

```python

if hasattr(model, 'intercept_'):
    print(f"모델의 절편 (intercept_): {model.intercept_}")

# 기타 모델의 속성 확인
# 필요한 속성들을 추가적으로 확인 가능
# 예를 들어, 의사결정 트리 모델의 경우 트리의 깊이 등을 확인할 수 있음
if hasattr(model, 'tree_'):
    print(f"트리의 깊이: {model.tree_.max_depth}")
    print(f"노드의 수: {model.tree_.node_count}")
```
